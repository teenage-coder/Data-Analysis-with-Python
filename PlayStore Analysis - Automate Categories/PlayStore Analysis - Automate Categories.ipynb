```python
import pandas as pd
```

```python
df = pd.read_csv("Dataset/googleplaystore.csv").dropna()
```

```python
df.head()
```

# Automatic Category Analysis

```python
unique_categories= list(set(df['Category']))
```

```python
unique_categories
```

```python
categories = df['Category']

temp = 0

for category in categories:
    if (category == "BUSINESS"):
        temp += 1
        
print(temp)
```

```python
for i in unique_categories:
    print(i)
```

```python
data = {}

categories = df['Category']


for category in unique_categories:

    temp = 0
    for i in categories:
        if (i == category):
            temp += 1
        
    data[category] = temp
```

# Automatic Genres Analysis

```python
unique_genres = list(set(df['Genres']))
```

```python
len(unique_genres)
```

```python
unique_genres
```

```python
data = {}

genres = df['Genres']


for genre in unique_genres:

    temp = 0
    for i in genres:
        if (i == genre):
            temp += 1
        
    data[genre] = temp
```
