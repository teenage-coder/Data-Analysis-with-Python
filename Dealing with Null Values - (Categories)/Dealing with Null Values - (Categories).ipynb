```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("Data.csv")
```

```python
df.iloc[ : , :]
```

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values = np.nan , strategy = 'median')
```

```python
imputer.fit(df.iloc[ : , 1:3])
```

```python
df
```

```python
df.iloc[ : , 1:3] = imputer.transform(df.iloc[ : , 1:3])
```

```python
df
```
