```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('USA_cars_datasets.csv')
```

```python
df.head()
```

## Q1. Maximum number of cars of every brand

```python
data = []

for brand in df['brand'].unique():
    data.append([brand , len(df[df['brand'] == brand])])
    
data = pd.DataFrame(data , columns = ['brand','freq'])
data = data.sort_values( by = 'freq' , ascending = False)

data.plot(x = 'brand', y = 'freq', kind = 'bar', figsize = (30,10))
```

## Q2 . Top 5 Car brands in terms of number of cars

```python
data = []

for brand in df['brand'].unique():
    data.append([brand , len(df[df['brand'] == brand])])
    
data = pd.DataFrame(data , columns = ['brand','freq'])
data = data.sort_values( by = 'freq' , ascending = False)

data = data.head()

data.plot(x = 'brand', y = 'freq', kind = 'bar', title = "Top 5 Car Brands in", grid = False, legend = False)
```

## Q3. Frequency of cars with respect to year

```python
data = []

for year in df['year'].unique():
    data.append([year , len(df[df['year'] == year])])
    
data = pd.DataFrame(data , columns = ['year','cars'])

data = data.sort_values( by = 'year', ascending = False)

data.plot(x = 'year' , y = 'cars' , kind = 'line' , title = 'Cars wrt Years')
```

```python
sns.violinplot(x = 'year', data = df, color = 'orange')
```

## Q4. Most Popular Car Colours

```python
data = []

for color in df['color'].unique():
    data.append([color , len(df[df['color'] == color])])
    
data = pd.DataFrame(data , columns = ['color','freq'])
data = data.sort_values( by = 'freq' , ascending = False)

data.plot(x = 'color', y = 'freq', kind = 'bar', figsize = (30,10))
```

## Q5. Top 10 Car Colours

```python
data = []

for color in df['color'].unique():
    data.append([color , len(df[df['color'] == color])])
    
data = pd.DataFrame(data , columns = ['color','freq'])
data = data.sort_values( by = 'freq' , ascending = False).head(10)

data.plot(x = 'color', y = 'freq', kind = 'bar', figsize = (20,10))
```

## Q7. Which car brand is having maximum average price?

```python
plt.figure( figsize = (20,10))
df.groupby('brand').mean()['price'].sort_values(ascending = False).plot.bar()
plt.tight_layout()
```

## Q8. Which state is having most expensive cars?

```python
plt.figure(figsize = (20,10))

df.groupby('state').mean()['price'].sort_values(ascending = False).plot.bar(color = 'brown')
```

```python
plt.figure(figsize = (10,5))

df.groupby('state').mean()['price'].sort_values(ascending = False).head(10).plot.bar(color = 'brown')
```
